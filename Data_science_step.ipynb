```python
#split into 3 types, each type will be processed separately
CATEGORICAL_COLUMN_NAMES=["make","fuel-type","aspiration","body-style",
                          "drive-wheels","engine-location",
                          "engine-type","fuel-system",]

CATEGORY_2_NUMBER_NAMES=["num-of-doors","num-of-cylinders"]

NUMBERIC_COLUMN_NAMES=["symboling","normalized-losses",
                       "wheel-base","length","width",
                       "height","curb-weight","engine-size",
                       "bore","stroke","compression-ratio",
                       "peak-rpm","city-mpg","highway-mpg"]
```

```python
# Make a copy first
pd_copy = df[df["price"] != "?"].copy()
pd_copy.head()
```

```python
#Process CATEGORI_2_NUMBER_NAMES
for col in CATEGORY_2_NUMBER_NAMES:
  #1st step. fill missing values by mode 
  highest_freq_value = pd_copy[col].mode()
  pd_copy.loc[pd_copy[col]=="?",col]=highest_freq_value
  #2nd step. convert string to number
  pd_copy[col] = pd_copy[col].map({"one":1,"two":2,
                                  "three":3,"four":4,
                                   "five":5,"six":6,
                                   "seven":7,"eight":8,
                                   "nine":9,"ten":10})
  #3rd. Add to NUMBERIC_COLUMN_NAMES
  NUMBERIC_COLUMN_NAMES.append(col)
```

```python
#Process NUMBERIC_COLUMN_NAMES
for col in NUMBERIC_COLUMN_NAMES:
  #1st step. fill missing values by mode 
  pd_copy.loc[pd_copy[col]=="?",col]=np.NaN
  pd_copy.loc[:,col].fillna(pd_copy[col].median(),inplace=True)
  pd_copy.loc[:,col] = pd_copy.loc[:,col].astype('float64')
  #print(pd_copy[col])
  max_col = np.max(pd_copy[col])
  mean_col = pd_copy[col].mean()
  pd_copy[col] = (pd_copy[col].subtract(max_col)).divide(mean_col)
```

```python
#Process categorical list
#1st step. Fill nan or ? with highest freq category
for col in CATEGORICAL_COLUMN_NAMES:
  highest_freq_value = pd_copy[col].mode()
  pd_copy.loc[pd_copy[col]=="?",col]=highest_freq_value

#2nd. OneHotEncoder 
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
cat_1hot = cat_encoder.fit_transform(pd_copy[CATEGORICAL_COLUMN_NAMES])
cat_1hot = cat_1hot.toarray()
```

```python
#Combine 2 data into training data
print(cat_1hot.shape)
print(pd_copy[NUMBERIC_COLUMN_NAMES].to_numpy().shape)
```

```python
X = np.concatenate((cat_1hot,pd_copy[NUMBERIC_COLUMN_NAMES].to_numpy()),axis=1)
print(X.shape)
```

```python
y = pd_copy["price"]
```

* Tree regression

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(max_depth=5)
tree_reg.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
ALL_LABELS=[]
ALL_LABELS= [item for sublist in cat_encoder.categories_ for item in sublist]
flatten_list = [item for sublist in cat_encoder.categories_ for item in sublist]
print(len(flatten_list))
for item in NUMBERIC_COLUMN_NAMES:
  ALL_LABELS.append(item)
#ALL_LABELS.append(item for item in NUMBERIC_COLUMN_NAMES)
print(len(ALL_LABELS))
export_graphviz(
tree_reg,
out_file="car_tree.dot",
feature_names=ALL_LABELS,
class_names="price",
rounded=True,
filled=True
)
```

```python
#Convert to png image from dot file
import pydot
(graph,) = pydot.graph_from_dot_file('car_tree.dot')
graph.write_png('car_tree.png')
```

* Gradient boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X, y)
```

```python
#calculate mean square error with training set
from sklearn.metrics import mean_squared_error
y_pred = gbrt.predict(X)
rmse = mean_squared_error(y_pred,y)
print("Mean square error:",rmse)
```

* Machine learning model

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([
keras.layers.Dense(X.shape[1], activation=tf.nn.relu),
keras.layers.Dense(128, activation=tf.nn.relu),
keras.layers.Dense(64, activation=tf.nn.relu),
keras.layers.Dense(32, activation=tf.nn.relu),
keras.layers.Dense(16, activation=tf.nn.relu),
keras.layers.Dense(1, activation=tf.nn.relu)
])
model.compile(optimizer=tf.optimizers.Adam(),
loss='mse')
print(type(X))
print(y.astype('float64').to_numpy())

model.fit(X, y.astype('float64').to_numpy(), epochs=2000)
```
